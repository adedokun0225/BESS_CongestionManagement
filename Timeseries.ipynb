```python
def clean_numeric_columns(df):
    for col in df.columns:
        # Only try to convert if column looks numeric
        if df[col].dtype == object:
            df[col] = (
                df[col]
                .astype(str)
                .str.strip()        # remove spaces
                .str.replace(",", ".")  # convert comma decimals
            )
        # Try convert to numeric, ignore errors for non-numeric columns
        df[col] = pd.to_numeric(df[col], errors="ignore")
    return df
```

```python
df_wind = clean_numeric_columns(df_wind)
df_photovoltaics = clean_numeric_columns(df_photovoltaics)
df_conventionalLoads = clean_numeric_columns(df_conventionalLoads)
df_electrodeHeater = clean_numeric_columns(df_electrodeHeater)
```

```python
filename= "timeseriesdata/wind_data.csv"
df_wind.to_csv(filename, index=False)
filename1= "timeseriesdata/photovoltaic_data.csv"
df_photovoltaics.to_csv(filename1, index=False)
filename2= "timeseriesdata/conventionLoad_data.csv"
df_conventionalLoads.to_csv(filename2, index=False)
filename3= "timeseriesdata/electrodeHeater.csv"
df_electrodeHeater.to_csv(filename3, index=False)

```
